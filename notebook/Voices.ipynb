```kotlin
@file:Repository("https://repo1.maven.org/maven2")
@file:DependsOn("org.pitest.voices:chorus:0.0.8")
@file:DependsOn("org.pitest.voices:alba:0.0.8")
@file:DependsOn("org.pitest.voices:en_uk:0.0.8")
@file:DependsOn("org.pitest.voices:g2p:0.0.8")
@file:DependsOn("com.microsoft.onnxruntime:onnxruntime:1.22.0")

import org.pitest.voices.Chorus
import org.pitest.voices.ChorusConfig
import org.pitest.voices.alba.Alba
import org.pitest.voices.uk.EnUkDictionary
```

```kotlin
// Verify that the G2P class is available on the classpath (should not throw)
Class.forName("org.pitest.voices.g2p.util.Resource")
println("g2p util.Resource found OK")
```

```kotlin
import java.nio.file.Paths

val config = chorusConfig(EnUkDictionary.en_uk())
val c = Chorus(config)
val alba = c.voice(Alba.albaMedium())
val audio = alba.say(
  """
        The term "London Times" is not the official name of the newspaper. The correct and official title is "The Times".
 It is a British daily newspaper based in London, established on January 1, 1785, under the name The Daily Universal Register, which was changed to The Times on January 1, 1788.
 The newspaper is often referred to as The London Times or The Times of London in countries where other publications with similar names are popular, but this is merely a descriptive reference rather than the official title.
      """.trimIndent()
)
val out = Paths.get("voices-test.wav")
audio.save(out)
audio.play()
```
